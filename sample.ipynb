```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11  # Gravitational constant in Nm^2/kg^2
M_sun = 1.989e30  # Mass of the Sun in kg

# Planet data
planet_names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]
planet_masses = np.array([3.285e23, 4.867e24, 5.972e24, 6.39e23, 1.898e27, 5.683e26, 8.681e25, 1.024e26])  # kg
planet_distances = np.array([5.791e10, 1.082e11, 1.496e11, 2.279e11, 7.785e11, 1.433e12, 2.877e12, 4.503e12])  # m

# Calculate gravitational force for each planet
forces = G * (M_sun * planet_masses) / (planet_distances**2)

# Display results
print("Gravitational Force of each Planet towards the Sun:")
for name, force in zip(planet_names, forces):
    print(f"{name}: {force:.2e} N")

# Plotting the gravitational force for each planet
plt.figure(figsize=(12, 6))  # Set figure size
plt.bar(planet_names, forces, color='mediumseagreen', edgecolor='black')

# Add labels, title, and grid
plt.title("Gravitational Force of Each Planet towards the Sun", fontsize=16)
plt.xlabel("Planet", fontsize=14)
plt.ylabel("Gravitational Force (N)", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show values on top of bars
for i, force in enumerate(forces):
    plt.text(i, force, f"{force:.2e}", ha='center', va='bottom', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
